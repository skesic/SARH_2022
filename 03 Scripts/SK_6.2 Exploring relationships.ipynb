#### Creating a categorical variable that splits the "Lydd" column into categories

```python
# Use a histogram to visualize the distribution of the variables.
# This way, you can determine sensible categories for the Rescue and Assisted taskings, by Base, ranges.
# The argument "kde" add a line that encompasses the distribution.

sns.histplot(sub_sarh0113_2['Lydd'], bins = 30, kde = True)
```

```python
sub_sarh0113_2.loc[sub_sarh0113_2['Lydd'] < 20, 'Rescue & Assisted taskings category'] = 'Highest number of taskings'
```

```python
sub_sarh0113_2.loc[(sub_sarh0113_2['Lydd'] >= 20) & (sub_sarh0113_2['Lydd'] <= 30), 'Rescue & Assisted taskings category'] = 'Average number of taskings'
```

```python
sub_sarh0113_2.loc[(sub_sarh0113_2['Lydd'] > 30) & (sub_sarh0113_2['Lydd'] < 60), 'Rescue & Assisted taskings category'] = 'Lowest number of taskings'
```
